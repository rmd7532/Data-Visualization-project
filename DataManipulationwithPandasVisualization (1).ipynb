```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 14, 8, 20, 12]

# Create a line plot
plt.plot(x, y, marker='o', linestyle='--', color='b', label='Line Plot')

# Adding labels and a legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot Example')
plt.legend()

# Show the plot
plt.show()
```

```python
# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [15, 24, 20, 30]

# Create a bar chart
plt.bar(categories, values, color='g', alpha=0.7, label='Bar Chart')

# Adding labels and a legend
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.legend()

# Show the plot
plt.show()
```
